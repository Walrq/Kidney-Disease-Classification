```python
import gdown 
```

```python
url = "https://drive.google.com/file/d/13yJVY6iQksPbPMDnx2WYbQ9bgcNlWYKQ/view?usp=sharing"
```

```python
file_id = url.split("/")[-2]
```

```python
prefix = 'https://drive.google.com/uc?/export=download&id='
gdown.download(prefix+file_id , "dataset.zip")
```
